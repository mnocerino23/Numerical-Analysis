```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

# 5a: Graph of f(x) on [-5 x 10^-5, 5 x 10^-5]:


```python
#set the x range from 0-pi (given by numpy pi) and use a step size of 0.001
x = np.arange(-5e-5,5e-5,1e-7)
#f(x) = numpy's sin(x)
y = x - (np.sin(x))

plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('f(x) = x-sin(x)')
```

# 5a: Graph of g(x) on [-5 x 10^-5, 5 x 10^-5]:





```python
import math as m

x = np.arange(-5e-5,5e-5,1e-7)
#f(x) = numpy's sin(x)
y = ((x**3)/(m.factorial(3))) - (x**5)/(m.factorial(5))

plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('g(x)')
```
