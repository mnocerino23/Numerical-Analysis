```python
import numpy as np
import math as m
import matplotlib.pyplot as plt


%matplotlib inline
```

# 5a: Graph of f(x) on [-5 x 10^-5, 5 x 10^-5]:


```python
#set the x range from -5 x 10^-5 to  5 x 10^5, and plot 1000 uniformly spaced points

x = np.arange(-5e-5,5e-5,1e-7)
y = x - (np.sin(x))

plt.title('f(x) = x-sin(x)')
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('f(x) = x-sin(x)')
```

# 5a: Graph of g(x) on [-5 x 10^-5, 5 x 10^-5]:





```python
#set the x range from -5 x 10^-5 to  5 x 10^5, and plot 1000 uniformly spaced points

x = np.arange(-5e-5,5e-5,1e-7)
y = ((x**3)/(m.factorial(3))) - (x**5)/(m.factorial(5))

plt.title('g(x): Taylor Series Approximation')
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('g(x)')
```
